```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
from pillow_heif import register_heif_opener


register_heif_opener()
```

```python
perms = ['000', '100', '110', '111', '101', '011', '010', '001']

for p in perms:
    df = pd.read_csv('data/sequences/3NDR/{}.csv'.format(p))
    pic = Image.open('data/sequences/3NDR/Photos/{}.heic'.format(p))
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(8,4))
#     for i in range(3):
#         ax1.plot(df['V{}'.format(i)])
    ax1.plot(df.V0)
    ax1.set_xlabel('Time [steps]')
    ax1.set_ylabel(r'$V_0$ [V]')
    ax2.imshow(pic)
    ax2.axis('off')
    fig.savefig('figures/sequences/3NDR/{}_V0sequence.png'.format(p))
    plt.show()
```

```python
df = pd.read_csv('data/sequences/3NDR/V0scan_updown.csv')

fig, (ax1,ax2) = plt.subplots(1,2, figsize=(10,4))
ax1.plot(df.V0, df.Vr/df.R0*1e3)
ax1.set_xlabel(r'$V_0$ [V]')
ax1.set_ylabel(r'$I$ [mA]')

for i in range(1,4):
    ax2.plot(df['V{}'.format(i)], label = r'$\Delta V{}$'.format(i))
ax2.legend()
ax2.set_xlabel('Time [steps]')
ax2.set_ylabel(r'NDR $\Delta V$ [V]')
fig.tight_layout()
fig.savefig('figures/sequences/3NDR/V0ramp_updown.png')
```
